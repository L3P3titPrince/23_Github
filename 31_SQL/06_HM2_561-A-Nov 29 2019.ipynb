1.For each customer, product and state combination, compute 
- (1) the customer's average sale of this product for the state, 
- (2) the average sale of the product and the customer but for the other states and 
- (3) the customer’s average sale for the given state, but for the other products.

```python
%%sql
with avg_table as(
    select cust, prod, state,round(avg(quant),2) as cust_avg 
    from sales
    group by cust, prod, state
),
avg_state as(
    select a.cust,a.prod,a.state, avg(b.quant) as other_state_avg
    from sales as a, sales as b
    where a.cust=b.cust
    and a.prod=b.prod
    and a.state != b.state
    group by a.cust,a.prod,a.state
),
avg_prod as(
    select a.cust,a.prod,a.state, avg(b.quant) as other_prod_avg
    from sales as a
    join sales as b
    on a.cust=b.cust
    and a.prod<>b.prod
    and a.state=b.state
    group by a.cust,a.prod,a.state
)
select distinct a.cust,a.prod,a.state,
round(a.cust_avg,2) as cust_avg,
round(b.other_state_avg,2) as other_state_avg,
round(c.other_prod_avg,2) as other_prod_avg
from avg_table as a
natural inner join avg_state as b 
natural inner join avg_prod as c
order by a.cust,a.prod
```

2.For customer and product, show the average sales before and after each quarter (e.g., for Q2, show average sales of Q1 and Q3. For “before” Q1 and “after” Q4, display <NULL>. The “YEAR” attribute is not considered for this query – for example, both Q1 of 2007 and Q1 of 2008 are considered Q1 regardless of the year.

```python
%%sql
with t_all_month as(
    select distinct s1.prod, s1.cust,
    case when s2.month in (1,2,3) then 1
         when s2.month in (4,5,6) then 2
         when s2.month in (7,8,9) then 3
         when s2.month in (10,11,12) then 4 
    end as q
    from sales s1, sales s2
    order by s1.prod,s1.cust
),
t1 as (
    select prod, cust, round(avg(quant)) as avq,
    case when month in (1,2,3) then 1
         when month in (4,5,6) then 2
         when month in (7,8,9) then 3
         when month in (10,11,12) then 4 
    end as q
    from sales
    group by prod, cust,q
),
t2 as(
    select t1.prod, t1.cust, bf.Q, bf.avq as bfq
    from t1
    left join t1 bf on t1.cust = bf.cust and t1.prod = bf.prod and t1.Q-1 = bf.q
),
t3 as(
    select t1.prod, t1.cust, af.Q, af.avq as afq
    from t1
    left join t1 af on t1.cust = af.cust and t1.prod = af.prod and t1.Q+1 = af.q
)
select ta.cust, ta.prod, ta.q,
bfq as before_avg, afq as after_avg
from t_all_month as ta
left join t1 on ta.prod = t1.prod and ta.cust = t1.cust and ta.q = t1.q  
left join t2 on t1.Q = t2.Q+1 AND t1.cust = t2.cust AND t1.prod = t2.prod 
left join t3 on t1.Q = t3.Q-1 AND t1.cust = t3.cust AND t1.prod = t3.prod
order by ta.cust, ta.prod, ta.q
```

3.For each product, find the median sales quantity (assume an odd number of sales for simplicity of presentation). (NOTE – “median” is defined as “denoting or relating to a value or quantity lying at the midpoint of a frequency distribution of observed values or quantities, such that there is an equal probability of falling above or below it.” E.g., Median value of the list {13, 23, 12, 16, 15, 9, 29} is 15.

```python
%%sql
SELECT
    prod,
    percentile_disc(0.5) WITHIN GROUP (ORDER BY quant) AS median_quant
FROM sales
GROUP BY prod
ORDER BY prod
```

```python
%%sql
WITH t1 AS(
SELECT t1.prod, t1.quant,
sum(case when t2.quant>= t1.quant then 1 else 0 end ) AS big,
sum(case when t2.quant<= t1.quant then 1 else 0 end ) AS sml,
-- count(t2.prod) as CNT,
ABS( sum(case when t2.quant>= t1.quant then 1 else 0 end ) -  sum(case when t2.quant<= t1.quant then 1 else 0 end ) ) as DIF

FROM sales t1, sales t2
WHERE t1.prod = t2.prod
GROUP BY 1,2
ORDER BY 1,2
),

t2 AS(
SELECT prod, min(DIF) as MIN_DIF, count(*)/2+1 as cnt
FROM t1
GROUP BY 1
)

SELECT t1.prod, t1.quant
FROM t1
INNER JOIN t2 ON t1.prod = t2.prod AND t1.big = t2.cnt  -- median number
ORDER BY 1
```
