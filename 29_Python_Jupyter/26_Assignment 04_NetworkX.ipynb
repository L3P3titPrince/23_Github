### 2.A paragraph describing the PageRank algorithm

PageRank algorithm is algorithm used by Google for getting a good searching result as a search engine. The basic method of PageRank is using "important" page to value other page. It's seems like a vote mechanism, very web page has its own "value". If one page have N outlink to other pages, one of the backline page will get 1/N value from this page. Bascially, more backline from other pages will increase the value of this page and this page will get more "votes" to get higher rank in internet.

Except this simplify model, PageRank also have severl extra method to help PageRank evaluate pages, for instance, Danling Links removable, adjust value to help newpage get equivalent value as older page,

According to PageRank Algothim, we find Node 14 is most important, it seems that degree_centrality agree with PageRank but betweeness centrality didn't agree. Because degree_centrality measured by neighbors effect, degree_centrality think node 14 is important beacuse it have most position in all neigbors connects, which is similar to some part of pagerank. On the other hand, betweenness_centrality and is influential in spreading information across a network, which means its value have no relationship with important postion in whole network. So betweenness_centrality chooese node as the most important

```python
#get the max value in PageRank and its value
print(type(pr_1))
max_key = max(pr_1, key=pr_1.get)
print(max_key,pr_1[max_key])
```

```python
#get max value in degree centrality and its value
dc=nx.degree_centrality(g_1)
dc_max=max(dc,key=dc.get)
print(dc_max,dc[dc_max])
```

```python
#get max value in betweenness_centrality and its value
bc=nx.betweenness_centrality(g_1)
bc_max=max(bc,key=bc.get)
print(bc_max,bc[bc_max])
```
